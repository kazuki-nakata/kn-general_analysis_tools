```python
import sys
import os
sys.path.append("..")
from knool.remote_sensing import modis
import glob
import numpy as np
import matplotlib.pyplot as plt
from osgeo import gdal
import knool.geodata_processor as gp
import time
#from knool.geodata_processor import geo_info
```

```python
#%%timeit
infile=r"../test_data/modis/MYD021KM.A2008246.1905.061.2018037220606.hdf"
dirname=os.path.dirname(infile); basename=os.path.basename(infile)
hdfpath2=glob.glob(dirname + os.sep + basename[0:3] + '03.' + basename.split('.')[1] + '*.hdf')[0]

raster=modis.Open(infile,hdfpath2,"em")
ist=raster.calc_IST()
raster.export_output(r"../test_data/output/myd021km_ist.tiff")
plt.imshow(ist)
```

```python
infile=r"../test_data/modis/MYD29.A2008246.1905.061.2021104190040.hdf"
raster=modis.Open(infile)
ist=raster.get_ist()
raster.export_output(r"../test_data/output/myd29_ist.tiff")
plt.imshow(ist)
```
